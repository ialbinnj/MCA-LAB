```python
import matplotlib.pyplot as plt 
import seaborn as sns 
import numpy as np 
import pandas as pd
```

```python
df = pd.read_csv("iris - iris.csv")
df
```

```python
sns.relplot(x='sepal.length',y='petal.length', data=df)
```

```python
sns.relplot(x='sepal.length',y='petal.length', data=df, hue='variety')
```

```python
sns.relplot(x='sepal.length',y='petal.length', data=df, hue='variety', style='sepal.length')
```

```python
df["sepal.length"].value_counts()
```

```python
sns.relplot(x='sepal.length',y='petal.length',kind='line', data=df)
```

```python
sns.relplot(x="sepal.length", y='petal.length', kind='line', data=df, height=5, aspect=4)
```

```python
sns.set(font_scale=2)
sns.relplot(x='sepal.length',y='petal.length',kind ='line', data=df, height=5, aspect=4)
```

```python
sns.set(font_scale=2)
sns.relplot(x='sepal.length',y='petal.length', kind='line', hue='variety', data=df, height=5, aspect=4)
plt.legend()
```
